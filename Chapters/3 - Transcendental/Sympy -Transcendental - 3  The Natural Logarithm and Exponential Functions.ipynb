```python
# Exponential
# Page 180 - Ch3 Calculus: A Complete Course

E3C = (sp.exp(x)-sp.exp(-x))/(sp.exp(x)+sp.exp(-x))
#E3C
D3C = E3C.diff()
D3C
simplify(D3C)
```
