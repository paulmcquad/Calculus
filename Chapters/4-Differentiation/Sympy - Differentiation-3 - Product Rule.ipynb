```python
# Calculcus - Differentiation - Page 111 - Ch2 Calculus: A Complete Course
# Differentiating gives the gradient of a graph

import sympy as sp
x,t = sp.symbols('x, t')
#f = exp(x)*cos(x)
#derivative_f = f.diff(x)
#derivative_f
```

```python
f1 = x**2 + 1 #Example 3 - Page 111 - Ch2 Calculus: A Complete Course
g1 = x**3+4
h1 = (f1*g1) # Product Rule
```

```python
EX3 = h1.diff(x)
EX3
```

```python
sp.simplify(EX3)
```

```python
from sympy import symbols
from sympy.plotting import plot

p1 = plot(EX3, title="Example 3")
```

```python
f2 = (2**sp.sqrt(x)+3/x) #Example 4 - Page 111 - Ch2 Calculus: A Complete Course
g2 = (3**sp.sqrt(x)-2/x)
h2 = (f2*g2) # Product Rule
```

```python
EX4 = h2.diff(x)
EX4
```

```python
sp.simplify(EX4)
```

```python
from sympy import symbols
from sympy.plotting import plot

p1 = plot(EX4, title="Example 4")
```
