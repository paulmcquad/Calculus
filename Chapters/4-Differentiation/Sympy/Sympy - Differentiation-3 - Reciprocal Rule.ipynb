```python
# Calculcus - Differentiation - Page 111 - Ch2 Calculus: A Complete Course
# Differentiating gives the gradient of a graph

import sympy as sp
x,t = sp.symbols('x, t')
```

```python
#Example 7a - Page 111 - Ch2 Calculus: A Complete Course

f1 = 1/(x**2+1) #Reciprocal Rule 
EX7A = f1.diff(x)
EX7A
```

```python
from sympy import symbols
from sympy.plotting import plot

p1 = plot(EX7A, title="Example 7(A)")
```

```python
#Example 7b - Page 111 - Ch2 Calculus: A Complete Course

f2 = 1/(t+(1/t)) #Reciprocal Rule 
EX7B = f2.diff(t)
EX7B
```

```python
from sympy import symbols
from sympy.plotting import plot

p2 = plot(EX7B, title="Example 7(B)")
```

```python
#Example 8 - Page 111 - Ch2 Calculus: A Complete Course

f3 = ((x**2+x+1)/x**3) #Reciprocal Rule 
EX8 = f3.diff(x)
sp.simplify(EX8)
```

```python
from sympy import symbols
from sympy.plotting import plot

p3 = plot(EX8, title="Example 8")
```
