```python
# Calculcus - Differentiation - Page 111 - Ch2 Calculus: A Complete Course
# Differentiating gives the gradient of a graph

import sympy as sp
a,b = sp.symbols('a, b')
m,n = sp.symbols('m, n')
x,t = sp.symbols('x, t')
theta = sp.Symbol("theta")
```

```python
#Example 9 (A) - Page 111 - Ch2 Calculus: A Complete Course

f1 = (1-x**2)/(1+x**2) #Quotient Rule 
EX9A = f1.diff(x)
sp.simplify(EX9A)
```

```python
from sympy import symbols
from sympy.plotting import plot

p1 = plot(EX9A, title="Example 9(A)")
```

```python
#Example 9 (B) - Page 111 - Ch2 Calculus: A Complete Course

f2 = sp.sqrt(t)/(3-5*t) #Quotient Rule 
EX9B = f2.diff(t)
#EX9B
sp.simplify(EX9B)
```

```python
from sympy import symbols
from sympy.plotting import plot

p2 = plot(EX9B, title="Example 9(B)")
```
