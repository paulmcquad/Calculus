```python
# Calculcus - Differentiation Trigonometric - Page 124 - Ch2 Calculus: A Complete Course
# Differentiating gives the gradient of a graph
from sympy import *
import sympy as sp
#a,b = sp.symbols('a, b')
#m,n = sp.symbols('m, n')
x,t = sp.symbols('x, t')
#theta = sp.Symbol("theta")
```

```python
#Example 2(A) Page 124 - Ch2 Calculus: A Complete Course

f2a = (sin(pi*x) + cos(3*x)) # Trigonometric 
EX2A = f2a.diff(x)
EX2A
```

```python
from sympy import symbols
from sympy.plotting import plot

p2a = plot(EX2A, title="Example 2(A)")
```

```python
#Example 2(B) Page 124 - Ch2 Calculus: A Complete Course

f2b = x**2*sin(sqrt(x)) # Trigonometric 
#f2b
EX2B = f2b.diff(x)
EX2B
#simplify(EX2B)
```

```python
from sympy import symbols
from sympy.plotting import plot

p2b = plot(EX2B, title="Example 2(B)")
```

```python
#Example 2(C) Page 124 - Ch2 Calculus: A Complete Course

f2c = (cos(x))/(1-sin(x)) # Trigonometric 
#2c
EX2C = f2c.diff(x)
#EX2C
simplify(EX2C)
```

```python
from sympy import symbols
from sympy.plotting import plot

p2c = plot(EX2C, title="Example 2(C)")
```
