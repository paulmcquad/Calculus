```python
# Calculcus - Differentiation - Page 116 - Ch2 Calculus: A Complete Course
# Differentiating gives the gradient of a graph

import sympy as sp
a,b = sp.symbols('a, b')
m,n = sp.symbols('m, n')
x,t = sp.symbols('x, t')
theta = sp.Symbol("theta")
```

```python
#Example 1 Page 116 - Ch2 Calculus: A Complete Course

f1 = (1)/(x**2-4) # Chain Rule 
EX1 = f1.diff(x)
EX1
```

```python
from sympy import symbols
from sympy.plotting import plot

p1 = plot(EX1, title="Example 1")
```

```python
#Example 2 Page 117 - Ch2 Calculus: A Complete Course

f2 = sp.sqrt(x**2+1) # Chain Rule
EX2 = f2.diff(x)
EX2
```
